```python
import cv2
import numpy as np
import matplotlib
from matplotlib.pyplot import *
%matplotlib inline
```

```python
image = cv2.imread('/Users/vitaly/Documents/Appls/2023/Anhalt/Slides/Figures/pirat.jpeg')
if image is None:
    print('Could not read image')
```

```python
imshow(image, cmap='gray', vmin=0, vmax=255)
axis('off')
```

```python
 
# Print error message if image is null
 
# Apply identity kernel
# kernel1 = np.array([[0, 0, 0],
#                     [0, 1, 0],
#                     [0, 0, 0]])

kernel1 = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])


# identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)
sobel_x = cv2.filter2D(src = image, ddepth=-1, kernel = kernel1)
imshow(sobel_x, cmap='gray', vmin=0, vmax=255)
axis('off')

# cv2.imshow('Original', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
kernel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])


# identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)
sobel_y = cv2.filter2D(src = sobel_x, ddepth=-1, kernel = kernel_y)
imshow(sobel_y, cmap='gray', vmin=0, vmax=255)
axis('off')
```

```python
kernel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])


# identity = cv2.filter2D(src=image, ddepth=-1, kernel=kernel1)
sobel_orig_y = cv2.filter2D(src = image, ddepth=-1, kernel = kernel_y)
imshow(sobel_orig_y, cmap='gray', vmin=0, vmax=255)
axis('off')
```
