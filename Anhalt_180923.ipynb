```python
import cv2
import numpy as np
import matplotlib
from matplotlib.pyplot import *
%matplotlib inline
```

# Importing image

```python
image = cv2.imread('pirat.jpeg')
if image is None:
    print('Could not read image')
```

```python
imshow(image, cmap='gray', vmin=0, vmax=255)
axis('off')
```

# Apply blurring kernel

```python
kernel2 = np.ones((5, 5), np.float32) / 25
img = cv2.filter2D(src = image, ddepth = -1, kernel = kernel2)
imshow(img, cmap='gray', vmin = 0, vmax = 255)
axis('off')
```

# Edge detection along x-Axis

```python
kernel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_x = cv2.filter2D(src = image, ddepth = -1, kernel = kernel_x)
imshow(sobel_x, cmap='gray', vmin=0, vmax=255)
axis('off')
```

# Edge detection along y-Axis

```python
kernel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])

sobel_y = cv2.filter2D(src = image, ddepth=-1, kernel = kernel_y)
imshow(sobel_y, cmap='gray', vmin=0, vmax=255)
axis('off')
```
